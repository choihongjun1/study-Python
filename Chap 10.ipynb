```python
df_1={'학번': ['1001','1002','1003'],
      '이름': ['홍길동','일지매','전우치'],
      '키' : [180,175,168] }
stu_1 = pd.DataFrame(df_1)
stu_1
```

```python
pd.merge(stu_1 , stu_2 , left_on='이름',right_on='이름')
```

```python
pd.merge(stu_1, stu_2, how='right',left_on='이름',right_on='이름')
```

```python
pd.merge(stu_1, stu_2, how='outer',left_on='이름',right_on='이름')
```

```python
from numpy import nan as NA
import pandas as pd
import numpy as np

member9 = {'번호' : [1,2,3,4,5,6,7] ,
           '이름' : ['홍길동','전우치',NA,'일지매','을지문덕',NA,'김유신'],
           '매출' : [100,200,300,NA,150,NA ,250]}
member10 = pd.DataFrame(member9 , columns=['번호','이름','매출'])
member10

print(member10.count())
print('NULL개수:',np.count_nonzero(member10['매출'].isnull()))
print('NaN개수:',np.count_nonzero(member10['이름'].isna()))
```

```python
import pandas as pd

member11 = {'번호' : [1,2,3,4,5] ,
            '팀명' : ['영업1팀','영업2팀','영업3팀','영업2팀','영업1팀'],
            '이름' : ['홍길동','전우치','일지매','을지문덕','홍길동'],
            '매출' : [100,200,300,150,250]}
member12 = pd.DataFrame(member11 , columns=['번호','팀명','이름','매출'])
member12
```

```python
group_12 =  member12['매출'].groupby(member12['팀명']) 
print('팀별매출합계:', group_12.sum() )
print()

group_12.sum =  member12['매출'].groupby([member12['팀명'],member12['이름']]).sum() 
print('팀별 이름별 매출합계:', group_12.sum )
print()

print('팀별 매출평균:', group_12.mean() )
print( )

print('팀별 매출건수:', group_12.count() )
```
